```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("/content/Iris.csv")
data.head()
```

```python
data["Species"].value_counts()
```

```python
sns.FacetGrid(data,hue = "Species",height = 6).map(plt.scatter,"PetalLengthCm","SepalWidthCm").add_legend()
```

```python
#Converting categorical variables to numbers
data['Species'].replace(['Iris-setosa', 'Iris-versicolor','Iris-virginica'],
                        [0, 1, 2], inplace=True)
```

```python
data.head()
```

```python
#Preparing the input and output variables
x = data[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]].values
y = data[["Species"]].values
```

```python
x
```

```python
y
```

```python
#Splitting the datasets
"""we are going to split the data into two datasets. One dataset is used for training the model and the other dataset is used for testing the model.
x_train,y_train are used for training purpose of the model and x_test,y_test are used for testing the model."""
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#logistic Regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
#Training the model
model.fit(x_train,y_train)
```

```python
#Testing the Accuracy of the model
model.score(x_train,y_train)
```

```python
#Make Predictions on new dataset which is not used while training the model.
expected = y_test
predicted = model.predict(x_test)
print(predicted)
expected = np.array(expected).flatten()
print(expected)
```

```python
dataset = pd.DataFrame({'Actual' : expected,"Predicted" : predicted})
dataset
#The model predicted with 100% accuracy without any wrong prediction.
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(expected,predicted))
#f1 score is the average of precision and recall of each class
#here 0 - represents Iris-setosa
#1 - represents Iris-versicolor
#2 - represents Iris-virginica
#so we got 100% accuracy when we tested it on new dataset which we have not used for training.
```

```python
print(metrics.confusion_matrix(expected,predicted))
"""Below is a 2Dimensional matrix in which diagonal elements represent the count
of each Species in the testing dataset i.e Iris-setosa = 11,Iris-versicolor = 13,Iris-virginica = 6"""
```
